```python
# Exmple of memory efficienecy
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Using the generator for Fibonacci sequence
fib_gen = fibonacci_generator()
for _ in range(10):
    print(next(fib_gen))
```

```python
# Exmple of memory Lazy Evaluation
def countdown(n):
    while n > 0:
        yield n
        n -= 1

# Using the generator for a countdown
countdown_gen = countdown(5)
for number in countdown_gen:
    print(number)
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
```

```python
# Test from 1 to 200
for n in range(1, 201):
    print(f"{n}: {'Prime' if is_prime(n) else 'Not Prime'}")
```

```python
#Fibonacci series>> at each stage the sum of last of two numbers takes place
def fib(n):
    a = 0
    b = 1
    for i in range (n):# if n is equal to 5 0.1 .2 3
        yield a
        a,b = b, a+b # assign the value of b to a then b is sum of a+b
```

```python
#code
def remove_v (s):
    vowels = 'aeiouAEIOU'
    result = ""
    for ch in s:
        if ch not in vowels:
            result += ch
    return result
```
